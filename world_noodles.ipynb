```python
import numpy as np
import pandas as pd
```

```python
url = "https://instantnoodles.org/en/noodles/demand/table/"
tables = pd.read_html(url)

# first table on the webpage
wina_df = tables[0]
wina_df.head()
```

```python
#little clean up
wina_df = wina_df.drop(wina_df.columns[0], axis=1)
wina_df.head()
```

```python
wina_df['Country/Region'] = wina_df['Country/Region'].replace(['China/ Hong Kong'], 'China')
wina_df['Country/Region'] = wina_df['Country/Region'].replace(['Republic of Korea'], 'South Korea')
```

```python
wina_df.head()
```

### Merge Population

```python
population_df = pd.read_csv('/content/world_population.csv')
population_df = population_df[['Rank', 'CCA3', 'Country/Territory', 'Capital', 'Continent', '2022 Population']]
population_df.head()
```

```python
wina_df = wina_df.merge(population_df, left_on='Country/Region', right_on='Country/Territory')
wina_df
```

```python
wina_df.to_csv("noodles.csv", index=False)
```
